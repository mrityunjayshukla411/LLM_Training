```python
from langchain_community.document_loaders import PyPDFLoader

file_path = ("/content/merged-pdf.pdf")
loader = PyPDFLoader(file_path)
pages = loader.load_and_split()
```

```python
len(pages)
```

```python
pages = pages[11:]
```

```python
  import pandas as pd
  import re
  import string

  def preprocess_text(text, max_length=512):
      """
      Preprocesses text data for feeding into a local LLM.

      Args:
          text (str): The input text data.
          max_length (int, optional): The maximum length of each text chunk. Defaults to 512.

      Returns:
          list: A list of preprocessed text chunks.
      """
      # Remove unnecessary whitespace
      text = re.sub(r'\s+', ' ', text).strip()

      # Remove URLs
      text = re.sub(r'http\S+', '', text)

      # Remove special characters
      text = text.translate(str.maketrans('', '', string.punctuation))

      # Convert to lowercase
      text = text.lower()

      # Split text into chunks
      chunks = [text[i:i+max_length] for i in range(0, len(text), max_length)]

      return chunks

  # Preprocess the text from each Document object
  preprocessed_texts = []
  for doc in pages:
      text = doc.page_content
      preprocessed_chunks = preprocess_text(text)
      preprocessed_texts.extend(preprocessed_chunks)

  # Create a DataFrame with the preprocessed text chunks
  df = pd.DataFrame({'text': preprocessed_texts})
```

```python
df.to_csv('train.csv', index=False)
```
